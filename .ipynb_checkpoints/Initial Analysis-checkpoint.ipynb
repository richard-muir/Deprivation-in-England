```python
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
deprivation_xl = pd.ExcelFile('2010 UK Deprivation.xls')
dep_df = deprivation_xl.parse('Health & Disability')
dep_df.rename(columns={'HEALTH DEPRIVATION AND DISABILITY SCORE':'DEPRIVATION SCORE','RANK OF HEALTH DEPRIVATION AND DISABILITY SCORE (where 1 is most deprived)':'DEPRIVATION RANK'}, inplace=True)
```

```python
dep_df.head()
```

```python
crime_xl = pd.ExcelFile('2010 UK Crime.xls')
crime_df = crime_xl.parse('Crime')
```

```python
crime_df.columns=('LSOA CODE','LA CODE','LA NAME','GOR CODE','GOR NAME','CRIME SCORE','CRIME RANK')
crime_df.head()
```

```python
plt.plot(crime_df['CRIME SCORE'], crime_df['CRIME RANK'])
```

```python
crime_dep = pd.merge(dep_df, crime_df, on=['LSOA CODE','LA CODE','LA NAME','GOR CODE','GOR NAME'])
```

```python
crime_dep.head()
```

```python
plt.scatter(crime_dep['DEPRIVATION SCORE'],crime_dep['CRIME SCORE'])
```

```python
education_xl = pd.ExcelFile('2010 UK Education.xls')
edu_df = education_xl.parse('Education Skills & Training')
edu_df.rename(columns={'EDUCATION SKILLS AND TRAINING SCORE':'EDUCATION SCORE',
                       'RANK OF EDUCATION SKILLS AND TRAINING SCORE (where 1 is most deprived)':'EDUCATION RANK'}, inplace=True)
```

```python
crime_dep_edu = pd.merge(crime_dep, edu_df, on=['LSOA CODE','LA CODE','LA NAME','GOR CODE','GOR NAME'])
```

```python
employ_xl = pd.ExcelFile('2010 UK Employment.xls')
emp_df = employ_xl.parse('Employment')
emp_df.rename(columns={'RANK OF EMPLOYMENT SCORE (where 1 is most deprived)':'EMPLOYMENT RANK'}, inplace=True)
```

```python
crime_dep_edu_emp = pd.merge(crime_dep_edu, emp_df, on=['LSOA CODE','LA CODE','LA NAME','GOR CODE','GOR NAME'])
```

```python
employ_xl = pd.ExcelFile('2010 UK Employment.xls')
emp_df = employ_xl.parse('Employment')
emp_df.rename(columns={'RANK OF EMPLOYMENT SCORE (where 1 is most deprived)':'EMPLOYMENT RANK'}, inplace=True)
```

```python
income_xl = pd.ExcelFile('2010 UK Income.xls')
inc_df = income_xl.parse('Income')
inc_df.rename(columns={'RANK OF INCOME SCORE (where 1 is most deprived)':'POOR RANK'}, inplace=True)
```

```python
data = pd.merge(crime_dep_edu_emp, inc_df, on=['LSOA CODE','LA CODE','LA NAME','GOR CODE','GOR NAME'])
```

```python
data.head()
```

```python
data.rename(columns={'EMPLOYMENT RANK':'UNEMPLOYMENT RANK','EMPLOYMENT SCORE':'UNEMPLOYMENT SCORE',
                     'EDUCATION RANK':'UNEDUCATION RANK','EDUCATION SCORE':'UNEDUCATION SCORE','INCOME SCORE':'POOR SCORE'}, inplace=True)
```

```python
rank_vars = [data['DEPRIVATION RANK'],data['CRIME RANK'],data['UNEDUCATION RANK'],data['UNEMPLOYMENT RANK'],data['POOR RANK']]
score_vars = [data['DEPRIVATION SCORE'],data['CRIME SCORE'],data['UNEDUCATION SCORE'],data['UNEMPLOYMENT SCORE'],data['POOR SCORE']]
```

```python
line_plot = plt.plot(score_vars[0])
```

```python
histogram = plt.hist(score_vars[0], bins=100)
```

```python
barchart = plt.bar(score_vars[0], rank_vars[0])
barchart
```

```python
plt.scatter(score_vars[0], score_vars[1])
```

```python
sns.jointplot(score_vars[0],score_vars[1])
```

```python
for i in range(len(score_vars)):
    for j in range(len(score_vars)):
        sns.jointplot(score_vars[i], score_vars[j], kind='hex')
```
